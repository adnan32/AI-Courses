```python
q1=covid1.groupby(["month","year","countriesAndTerritories","popData2020"]).sum("cases")
q1=q1.reset_index()
q1=q1.sort_values(by=["year", "month","cases"],ascending=False)
qq1=pd.DataFrame({"month": [], "year": [], "countriesAndTerritories": [],"cases":[],"day":[],"deaths":[],"popData2020":[]})
for i in range(13):
    mont20=q1.loc[(q1["month"]==i)&(q1["year"]==2020)]
    #mont20=mont20.drop_duplicates(subset=["countriesAndTerritories","month","year"],keep="first")
    mont21=q1.loc[(q1["month"]==i)&(q1["year"]==2021)]
    #mont21=mont21.drop_duplicates(subset=["countriesAndTerritories","month","year"],keep="first")
    mont22=q1.loc[(q1["month"]==i)&(q1["year"]==2022)]
    #mont22=mont22.drop_duplicates(subset=["countriesAndTerritories","month","year"],keep="first")
    mo20=mont20.head(10)
    #print(mo20)
    qq1=qq1.append(mo20, ignore_index=True)
    mo21=mont21.head(10)
    qq1=qq1.append(mo21, ignore_index=True)
    mo22=mont22.head(10)
    qq1=qq1.append(mo22, ignore_index=True)
qq1
```

```python
month=pd.to_datetime(qq1["month"], format='%m')
month=month.dt.strftime("%B")
qq1["month"]=month
qq1=qq1.sort_values(by=["year","month"])
qq1
```

### 2. Visualize the total number of cases and deaths in each country on a map using the geographical locations in 2020-2021.

```python
country_deaths=covid1.loc[(covid1["year"]==2020)|(covid1["year"]==2021)]
country_deaths=covid1.groupby("countriesAndTerritories")["deaths"].sum()
```

```python
country_deaths=country_deaths.reset_index()
country_deaths
```

```python
country_cases=covid1.groupby("countriesAndTerritories")["cases"].sum()
```

```python
country_cases=country_cases.reset_index()
country_cases
```

```python
worldmap = [
    dict(
        type="choropleth",
        locations=country_cases["countriesAndTerritories"],
        locationmode="country names",
        z=country_cases["cases"],
        autocolorscale=True,
        reversescale=False,
        marker=dict(line=dict(color="rgb(180,180,180)", width=0.5)),
        colorbar=dict(autotick=False, title="Number of cases"),
    )
]

layout = dict(
    title="The Nationality of cases",
    geo=dict(showframe=False, showcoastlines=True, projection=dict(type="Mercator")),
)

fig = dict(data=worldmap, layout=layout)
py.iplot(fig, validate=False)
```

```python
worldmap = [
    dict(
        type="choropleth",
        locations=country_deaths["countriesAndTerritories"],
        locationmode="country names",
        z=country_deaths["deaths"],
        autocolorscale=True,
        reversescale=False,
        marker=dict(line=dict(color="rgb(180,180,180)", width=0.5)),
        colorbar=dict(autotick=False, title="Number of deeaths"),
    )
]

layout = dict(
    title="The Nationality of deaths",
    geo=dict(showframe=False, showcoastlines=True, projection=dict(type="Mercator")),
)

fig = dict(data=worldmap, layout=layout)
py.iplot(fig, validate=False)
```
