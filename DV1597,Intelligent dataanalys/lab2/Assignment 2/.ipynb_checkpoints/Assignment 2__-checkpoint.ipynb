```python
#vi ser att deaths missing value är antigen inte fyllda eller att det ingen som daog under den dagen därför fyllde de inte.
#deth & cases
#vi skapar en skala för att bestämma om vad vi ska göra med missing value:
#om vi har mer än fem dagar under samma månad och land som har missing values så replesar det med noll,
#annars ersätter vi det med månadens mean value.

# i våra bedömning kollar om mer än fem value är missade så kollar andra kolumner om i den andra cloumnen ser höga tal, 
# bedömmer vi om vi ska ta mean av månaden 

# vi tar den föra månadens mean och månadens efter mean och sen mean av de två.

covid1[(covid1['cases'].isnull())|(covid1['deaths'].isnull())]
```

```python
#check if there is more than one empty field in all records
covid1.loc[(covid1['cases'].isnull()) & (covid1['deaths'].isnull())]
```

```python
#using interpolation to fill missing values is possibale in case not founding any record from command above


geoId=list(covid1['geoId'].unique())
countries = list(covid1['countriesAndTerritories'].unique())
covid2 = pd.DataFrame({})
for country in countries:
    covid2 = covid2.append(covid1.loc[(covid1['countriesAndTerritories'] == country)].interpolate())
covid2.isnull().sum()
```

```python
#we found five null values that was not taken care of by interpolate so we take these values and give them the mean of the 5 previous records

nan = [1576,3914,5520,17309,21232]

for i in nan:
    if i == 3914:
        covid2.at[i,'deaths'] = round(covid2['deaths'].iloc[i-5:i].mean())
    else:
        covid2.at[i,'cases'] = round(covid2['cases'].iloc[i-5:i].mean())
covid2.isnull().sum()
```

```python
#convert negative values to positive
covid2['cases'] = covid2['cases'].abs()
covid2['deaths'] = covid2['deaths'].abs()
# covid2 = covid2.loc[(covid2['cases'] < 10000)|(covid2['deaths']  < 500)]
covid2.boxplot(column='cases', figsize = [10,15])
# st.probplot(covid2['cases'], plot=plt)
# plt.show()
```

```python
#check for duplicates
covid2.duplicated(subset=['dateRep', 'countriesAndTerritories'], keep='first').sum()
```

```python
q1=covid2.groupby(["month","year","countriesAndTerritories","popData2020"]).sum("cases")
q1=q1.reset_index()
q1=q1.sort_values(by=["year", "month","cases"],ascending=False)
qq1=pd.DataFrame({"month": [], "year": [], "countriesAndTerritories": [],"cases":[],"day":[],"deaths":[],"popData2020":[]})
for i in range(13):
    mont20=q1.loc[(q1["month"]==i)&(q1["year"]==2020)]
    #mont20=mont20.drop_duplicates(subset=["countriesAndTerritories","month","year"],keep="first")
    mont21=q1.loc[(q1["month"]==i)&(q1["year"]==2021)]
    #mont21=mont21.drop_duplicates(subset=["countriesAndTerritories","month","year"],keep="first")
    mont22=q1.loc[(q1["month"]==i)&(q1["year"]==2022)]
    #mont22=mont22.drop_duplicates(subset=["countriesAndTerritories","month","year"],keep="first")
    mo20=mont20.head(10)
    #print(mo20)
    qq1=qq1.append(mo20, ignore_index=True)
    mo21=mont21.head(10)
    qq1=qq1.append(mo21, ignore_index=True)
    mo22=mont22.head(10)
    qq1=qq1.append(mo22, ignore_index=True)
qq1
```

```python
month=pd.to_datetime(qq1["month"], format='%m')
month=month.dt.strftime("%B")
qq1["month"]=month
qq1=qq1.sort_values(by=["year","month"])
qq1
```

#### Do you see any connection among these countries regarding their geographical locations and the month the most cases are observed?

the month there the most cases are observed is 

```python
countryname=qq1.drop_duplicates(subset=["countriesAndTerritories"])
max_co=pd.DataFrame({})
for i in countryname["countriesAndTerritories"]:
    c=qq1.loc[qq1["countriesAndTerritories"]==f"{i}"]
    row=c["cases"].max()
    row=c.loc[c["cases"]==row]
    max_co=max_co.append(row, ignore_index=True)
max_co   
```

```python
worldmap = [
    dict(
        type="choropleth",
        locations=max_co["countriesAndTerritories"],
        locationmode="country names",
        z=max_co["cases"],
        text=max_co["month"],
        autocolorscale=True,
        reversescale=False,
        marker=dict(line=dict(color="rgb(180,180,180)", width=0.5)),
        colorbar=dict(autotick=False, title="Number of cases"),
    )
]

layout = dict(
    title="The Nationality of cases",
    geo=dict(showframe=False, showcoastlines=True, projection=dict(type="Mercator")),
)

fig = dict(data=worldmap, layout=layout)
py.iplot(fig, validate=False)
```

### 3. What are top-3 popular vaccine brands that have been used across EU/EEA? Group the countries based on the vaccine brands and report your findings.

```python
vaccin=pd.read_csv("COVID-19_vaccination.csv")
vaccin.head(15)
```

```python
vaccin.shape
```

```python
vaccin.isnull().sum()
```

```python
vaccinTop3=vaccin.groupby(["Vaccine"])["FirstDose","SecondDose","UnknownDose","DoseAdditional1"].sum()
vaccinTop3["Total"]=vaccinTop3.FirstDose + vaccinTop3.SecondDose + vaccinTop3.UnknownDose + vaccinTop3.DoseAdditional1
vaccinTop3=vaccinTop3.sort_values(by="Total",ascending=False)
vaccinTop3.head(3)
```

```python
vaccin1=vaccin.groupby(["ReportingCountry","YearWeekISO","Vaccine"])["FirstDose","SecondDose","UnknownDose","DoseAdditional1"].sum()
vaccin1=vaccin.reset_index(drop=True)
vaccin1
```

```python
coun_vacc=vaccin1
coun_vacc.drop(coun_vacc.index[(coun_vacc['FirstDose'] == 0) & (coun_vacc["SecondDose"]==0) & (coun_vacc["UnknownDose"]==0) & (coun_vacc["DoseAdditional1"]==0)], inplace=True)
coun_vacc
```

```python
coun_vacc1=coun_vacc.drop_duplicates(subset=["ReportingCountry","Vaccine"])
coun_vacc1
```

```python
vaccine_count=coun_vacc1.groupby(["ReportingCountry"])["Vaccine"].count().reset_index()
country=['Austria','Belgium','Bulgaria','Cyprus','Czechia','Germany','Denmark','Estonia','Greece','Spain','Finland','France','Croatia','Hungary','Ireland','Iceland','Italy','Liechtenstein','Lithuania','Luxembourg','Latvia','Malta','Netherlands','Norway','Poland','Portugal','Romania','Sweden','Slovenia','Slovakia']
```

```python
plot_df=coun_vacc1.groupby(["ReportingCountry"]).agg({"Vaccine": ', '.join}).reset_index()
plot_df=plot_df.sort_values("ReportingCountry")
plot_df["contry"]=country
plot_df["Vaccin count"]=vaccine_count["Vaccine"]
plot_df
```

```python
worldmap = [
    dict(
        type="choropleth",
        locations=plot_df["contry"],
        locationmode="country names",
        z=plot_df["Vaccin count"],
        text=plot_df["Vaccine"],
        autocolorscale=True,
        reversescale=False,
        marker=dict(line=dict(color="rgb(180,180,180)", width=0.5)),
        colorbar=dict(autotick=False, title="Number of Vaccines"),
    )
]

layout = dict(
    title="The Nationality of Vaccines",
    geo=dict(showframe=False, showcoastlines=True, projection=dict(type="Mercator")),
)

fig = dict(data=worldmap, layout=layout)
py.iplot(fig, validate=False)
```
